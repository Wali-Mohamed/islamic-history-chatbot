```python
def find_headers(example):
    text = example['english'].lower()
    return "here ends" in text or text.startswith("the book of")

headers = ds['train'].filter(find_headers)

print(f"--- Found {len(headers)} Book/Chapter Markers ---")
for i, row in enumerate(headers.select(range(min(10, len(headers))))):
    print(f"{i+1}: {row['english']}")
```
