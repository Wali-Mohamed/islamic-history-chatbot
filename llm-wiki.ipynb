```python
import wikipediaapi
import os

# Initialize Wiki API
user_agent = "GenAI_Project_Bot/1.0 (academic_research_contact@example.com)"
wiki = wikipediaapi.Wikipedia(user_agent, 'en')

# The 4 key pages for GenAI
pages = [
    "Generative artificial intelligence",
    "Large language model",
    "Transformer (deep learning architecture)",
    "Diffusion model"
]

# Create data directory
if not os.path.exists('wiki_pages'):
    os.makedirs('wiki_pages')

for page_name in pages:
    page = wiki.page(page_name)
    if page.exists():
        filename = f"wiki_pages/{page_name.replace(' ', '_')}.md"
        with open(filename, "w", encoding="utf-8") as f:
            f.write(f"# {page.title}\n\n{page.text}")
        print(f"Successfully extracted: {page_name}")
    else:
        print(f"Error: Page '{page_name}' not found.")
```
