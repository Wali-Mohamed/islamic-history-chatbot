```python
# library
import pandas as pd
```

```python
df = pd.read_excel("Bananas Sales.xlsx",sheet_name=1, skiprows=0)
print(df.head())
```

```python
df.head()
```

```python
df["Sales per Sub"] = df["Sales Amount"] / df["Licenses Bought"]
```

```python
pivot = pd.pivot_table(
    df,
    index="Subscription Type",
    values=["Sales Amount", "Licenses Bought"],
    aggfunc= { "Sales Amount": ["sum", "mean"],
            "Licenses Bought": "sum" }
)
```

```python
pivot
```

```python
pivot.columns = ["_".join(col) for col in pivot.columns]
```

```python
pivot.columns
```

```python
pivot
```

```python
pivot["Sales per Sub"] = pivot["Sales Amount_sum"] / pivot["Licenses Bought_sum"]
```
